```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 读取数据
land_info = pd.read_csv("land_info.csv")
crop_info = pd.read_csv("crop_info.csv")
planting_2023 = pd.read_csv("planting_2023.csv")
stats_2023 = pd.read_csv("stats_2023.csv")

# 地块类型分布
plt.figure(figsize=(10, 6))
sns.countplot(data=land_info, x='地块类型')
plt.title('地块类型分布')
plt.xticks(rotation=45)
plt.savefig('地块类型分布.png')
plt.show()

# 作物类型分布
plt.figure(figsize=(10, 6))
sns.countplot(data=crop_info, x='作物类型')
plt.title('作物类型分布')
plt.xticks(rotation=45)
plt.savefig('作物类型分布.png')
plt.show()

# 亩产量分布
plt.figure(figsize=(10, 6))
sns.histplot(data=stats_2023, x='亩产量/斤', bins=20, kde=True)
plt.title('亩产量分布')
plt.savefig('亩产量分布.png')
plt.show()

# 检查数据缺失值
print("地块信息缺失值检查:")
print(land_info.isnull().sum())
print("\n作物信息缺失值检查:")
print(crop_info.isnull().sum())
print("\n2023年种植情况缺失值检查:")
print(planting_2023.isnull().sum())
print("\n2023年统计数据缺失值检查:")
print(stats_2023.isnull().sum())
```
